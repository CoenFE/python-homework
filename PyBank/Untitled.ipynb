```python
#import the pathlib and csv library 
from pathlib import Path 
import csv
```

```python
#setting file path
csv_path = Path("budget_data.csv")

#setting variables 
total_months = 0
month_of_change = []
net_change_list = []
greatest_increase = ["", 0]
greatest_decrease = ["", 9999999999999999999]
total_net = 0
sum_net_change = 0

#opening csv
with open(csv_path) as f:
    reader = csv.reader(f)
    #print(row)
    header = next(reader)
    first_row = next(reader)
    previous_net = int(first_row[1])
    for row in reader:
        print(previous_net)
        total_months = total_months + 1
        total_net = total_net + int(row[1])
        net_change = int(row[1]) - previous_net
        previous_net = int(row[1])
        sum_net_change = net_change + sum_net_change
        print(sum_net_change)
        if net_change > greatest_increase[1]:
            greatest_increase[0] = row[0]
            greatest_increase[1] = net_change
        #calculating the greatest decrease
        if net_change < greatest_decrease[1]:
            greatest_decrease[0] = row[0]
            greatest_decrease[1] = net_change
```

```python
#calculating the average change 
net_monthly_avg = round(sum_net_change / (total_months), 2)
print(net_monthly_avg)
```

```python
#output results to txt file
with open("./OUTPUT.txt", "w") as f:
    f.write(f"Financial Analysis\n")
    f.write(f"------------------------------------\n")
    f.write(f"Total Months: {total_months}\n")
    f.write(f"Total: ${total_net}\n")
    f.write(f"Average Change: ${net_monthly_avg}\n")
    f.write(f"Greatest Increase: ${greatest_increase}\n")
    f.write(f"Greatest Descrease: ${greatest_decrease}\n")
    
```
